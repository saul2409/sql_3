## **<font color="#000000"> Qu es SQL?</font>**
</td>
</tr>
</table>

**SQL** (Structured Query Language) es un lenguaje de programacin diseado para gestionar y manipular bases de datos relacionales.

**SQL** es fundamental en la gestin de datos, ya que permite a las organizaciones almacenar, recuperar y manipular informacin de manera efectiva y segura.

Es ampliamente utilizado en sistemas de gestin de bases de datos como MySQL, PostgreSQL, Microsoft SQL Server y Oracle.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Caractersticas </font>**
</td>
</tr>
</table>

>* **Declarativo**:
  * SQL es un lenguaje declarativo, lo que significa que los usuarios especifican qu datos desean obtener o manipular, sin necesidad de definir cmo se deben realizar estas operaciones.
>* **Estndar**:
  * SQL es un estndar reconocido por ANSI (American National Standards Institute) e ISO (International Organization for Standardization), lo que permite su uso en diversas plataformas de bases de datos.
>* **Interoperabilidad**:
  * Puede ser utilizado en diferentes sistemas de gestin de bases de datos, como MySQL, PostgreSQL, Microsoft SQL Server, Oracle, entre otros.

## **<font color="#000000"> Componentes Principales</font>**
</td>
</tr>
</table>

>* **Esquemas**: Proveen una forma de organizar las tablas y otros objetos dentro de la base de datos.
>* **Tablas**: Son la estructura bsica donde se almacenan los datos. Cada tabla consiste en filas y columnas.
>* **Consultas**: Se utilizan para realizar operaciones sobre los datos, como seleccionar, insertar, actualizar y eliminar registros.
>* **ndices**: Mejoran la velocidad de las consultas al permitir un acceso ms rpido a los datos.
>* **Vistas**: Son consultas guardadas que presentan datos de una o varias tablas de manera especfica. Actan como tablas virtuales y pueden simplificar la complejidad de las consultas.

## **<font color="#000000"> Esquemas</font>**
</td>
</tr>
</table>

Un esquema es una forma de organizar y agrupar objetos de base de datos, como tablas, vistas, procedimientos almacenados, funciones y ms.

Los esquemas proporcionan una estructura lgica que permite administrar mejor los objetos y gestionar la seguridad y el acceso a los mismos.

GvrHG.gif

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Caractersticas del Esquema</font>**
</td>
</tr>
</table>

>* **Agrupacin de Objetos**:
  * Los esquemas permiten agrupar objetos relacionados en una base de datos, lo que facilita la organizacin y la navegacin.
>* **Seguridad**:
  * Se pueden definir permisos a nivel de esquema, lo que permite controlar el acceso a todos los objetos dentro de un esquema de manera centralizada.
  * Por ejemplo, se puede otorgar acceso a un esquema especfico a ciertos usuarios, limitando su capacidad para acceder a otros objetos.
>* **Nombres nicos**:
  * Un objeto dentro de un esquema debe tener un nombre nico en el contexto del esquema, pero puede haber objetos con el mismo nombre en diferentes esquemas.
  * Esto permite una mayor flexibilidad en la nomenclatura de los objetos.
>* **Facilidad de Mantenimiento**:
  * Los esquemas facilitan el mantenimiento de objetos relacionados, ya que se pueden manejar como un conjunto.
  * Tambin pueden simplificar la migracin de objetos entre bases de datos.
