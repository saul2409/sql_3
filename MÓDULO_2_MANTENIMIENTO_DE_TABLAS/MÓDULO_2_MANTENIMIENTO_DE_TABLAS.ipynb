### **<font color="#FFFFFF"> Table Scan (Escaneo de tabla)</font>**
</td>
</tr>
</table>

>* **Descripcin**:
  * Un **table scan** ocurre cuando SQL Server lee todas las filas de una tabla para encontrar las que cumplen con una condicin especificada en la consulta.
>* **Cundo Ocurre**: Se utiliza generalmente cuando:
  * No hay ndices aplicables a la consulta.
  * La consulta solicita un porcentaje alto de las filas de la tabla.
  * Se est utilizando una tabla pequea, donde el costo de un escaneo completo es menor que el uso de un ndice.
>* **Desventajas**:
  * Puede ser ineficiente en tablas grandes, ya que requiere leer cada fila, lo que puede llevar mucho tiempo y recursos.

### **<font color="#FFFFFF"> Clustered Index (ndice Agrupado)</font>**
</td>
</tr>
</table>

>* **Descripcin**:
  * Un clustered index organiza fsicamente los datos en la tabla segn el orden del ndice. Cada tabla puede tener solo un ndice agrupado.
>* **Cundo Ocurre**:
  * Se utiliza para acceder a filas de manera rpida y eficiente, ya que los datos estn almacenados en el disco en un orden especfico.
>* **Ventajas**:
  * Mejora significativamente el rendimiento de las consultas que buscan rangos de datos o utilizan operaciones de ordenamiento.
  * Reduce la necesidad de leer filas innecesarias.

## **<font color="#000000"> Clave primaria (PK)</font>**
</td>
</tr>
</table>

>* Primary Key
>* Clave primaria
>* Clave principal
>* Llave primaria
>* Llave principal

Una clave primaria en SQL Server es un conjunto de una o ms columnas que identifican de manera nica cada fila en una tabla.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Caractersticas de la Clave Primaria</font>**
</td>
</tr>
</table>

>* **Unicidad**: Los valores en una clave primaria deben ser nicos. No puede haber dos filas con el mismo valor en la columna (o combinacin de columnas) que define la clave primaria.
>* **No Nulo**: Las columnas que forman la clave primaria no pueden contener valores nulos. Esto asegura que cada fila pueda ser identificada de manera nica.
>* **ndice**: Al definir una clave primaria, SQL Server crea automticamente un ndice nico en las columnas de la clave primaria. Esto mejora el rendimiento de las consultas que buscan filas especficas.
>* **Identificacin**: La clave primaria es utilizada para establecer relaciones entre tablas. Por ejemplo, puede ser referenciada por claves forneas en otras tablas, ayudando a mantener la integridad referencial.

## **<font color="#000000"> Identity (Valor_inicial, incremento)</font>**
</td>
</tr>
</table>

El trmino IDENTITY en SQL Server se utiliza para definir una columna que generar automticamente valores nicos y secuenciales para cada nuevo registro que se inserte en una tabla.

Esto es especialmente til para crear claves primarias.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Como funciona el IDENTITY? </font>**
</td>
</tr>
</table>

>* **Generacin Automtica**:
Cuando se define una columna como IDENTITY, SQL Server asigna automticamente un valor a esta columna cada vez que se agrega una nueva fila. No es necesario especificar un valor para esta columna en la instruccin INSERT.
>* **Valores Secuenciales**:
Los valores generados por IDENTITY son secuenciales, lo que significa que el primer valor es el que se define al crear la tabla (por defecto 1) y se incrementa en un valor especificado (por defecto 1) cada vez que se inserta una nueva fila.



```
USE empresa
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in INT IDENTITY
	CONSTRAINT pk_personal_rol_cod_rol_in
	PRIMARY KEY CLUSTERED(cod_rol_in)
	NOT NULL,
	cod_rol_ch CHAR(5),
	nom_rol_vc VARCHAR(20),
)
GO
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0001', 'Ventas');
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0002', 'Almacn');
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0003', 'Personal');
GO
SELECT *
FROM Personal.Rol
GO
```

## **<font color="#000000"> Clave alterna (AK)</font>**
</td>
</tr>
</table>

Una clave alterna es un conjunto de una o ms columnas que se utilizan para identificar de manera nica cada fila en una tabla, pero que no se designa como la clave primaria.

En otras palabras, una clave alterna es una opcin alternativa para identificar filas que tambin cumple con la propiedad de unicidad.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Caractersticas de la Clave Alterna</font>**
</td>
</tr>
</table>

>* **Unicidad**: Al igual que la clave primaria, una clave alterna debe contener valores nicos para cada fila en la tabla. No puede haber duplicados.
>* **No Nulidad**: Las columnas que forman una clave alterna no pueden contener valores nulos, asegurando que cada fila pueda ser identificada de manera nica.
>* **ndice nico**: Al definir una clave alterna, SQL Server puede crear un ndice nico en las columnas que la componen, lo que mejora el rendimiento de las consultas que utilizan estas columnas para buscar datos.
>* **Uso en Relaciones**: Aunque no es la clave primaria, una clave alterna puede ser referenciada por otras tablas, similar a cmo se hace con la clave primaria.



```
USE empresa
GO
IF EXISTS
(
    SELECT *
    FROM sys.schemas s
    INNER JOIN sys.tables t
    ON s.schema_id = t.schema_id
    WHERE
            s.name = 'Personal'
        AND
            t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
    cod_rol_in INT IDENTITY
    CONSTRAINT pk_personal_rol_cod_rol_in
    PRIMARY KEY CLUSTERED(cod_rol_in) NOT NULL,
    
	cod_rol_ch CHAR(5) UNIQUE,
    nom_rol_vc VARCHAR(20),
)
GO
SELECT *
FROM Personal.Rol
GO
```

### **<font color="#FFFFFF"> Tipos de ndices</font>**
</td>

>* **ndices Non-Clustered**: Son ndices separados de los datos de la tabla. Pueden existir mltiples ndices non-clustered en una tabla.
>* **ndices Compuestos**: Son ndices que incluyen ms de una columna. Se utilizan para consultas que filtran o buscan en mltiples columnas.
