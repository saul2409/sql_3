</td>
</tr>

En SQL Server, el acceso a los datos puede realizarse de diferentes maneras, siendo dos de las más comunes el **table scan** y el uso de **índices**, como el **clustered index**.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Table Scan (Escaneo de tabla)</font>**
</td>
</tr>
</table>

>* **Descripción**:
  * Un **table scan** ocurre cuando SQL Server lee todas las filas de una tabla para encontrar las que cumplen con una condición especificada en la consulta.
>* **Cuándo Ocurre**: Se utiliza generalmente cuando:
  * No hay índices aplicables a la consulta.
  * La consulta solicita un porcentaje alto de las filas de la tabla.
  * Se está utilizando una tabla pequeña, donde el costo de un escaneo completo es menor que el uso de un índice.
>* **Desventajas**:
  * Puede ser ineficiente en tablas grandes, ya que requiere leer cada fila, lo que puede llevar mucho tiempo y recursos.



```
USE empresa
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in INT,
	cod_rol_ch CHAR(5),
	nom_rol_vc VARCHAR(20),
)
GO
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (5, 'R0005', 'Ventas');
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (2, 'R0002', 'Almacén');
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (8, 'R0008', 'Personal');
GO
SELECT *
FROM Personal.Rol
GO
```

### **<font color="#FFFFFF"> Clustered Index (Índice Agrupado)</font>**
</td>
</tr>
</table>

>* **Descripción**:
  * Un clustered index organiza físicamente los datos en la tabla según el orden del índice. Cada tabla puede tener solo un índice agrupado.
>* **Cuándo Ocurre**:
  * Se utiliza para acceder a filas de manera rápida y eficiente, ya que los datos están almacenados en el disco en un orden específico.
>* **Ventajas**:
  * Mejora significativamente el rendimiento de las consultas que buscan rangos de datos o utilizan operaciones de ordenamiento.
  * Reduce la necesidad de leer filas innecesarias.



```
USE empresa
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in INT PRIMARY KEY,
	cod_rol_ch CHAR(5),
	nom_rol_vc VARCHAR(20),
)
GO
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (5, 'R0005', 'Ventas');
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (2, 'R0002', 'Almacén');
INSERT INTO Personal.Rol(cod_rol_in, cod_rol_ch, nom_rol_vc) VALUES (8, 'R0008', 'Personal');
GO
--SET SHOWPLAN_ALL ON
GO
SELECT *
FROM Personal.Rol
GO
--SET SHOWPLAN_ALL OFF
GO
```

## **<font color="#000000"> Clave primaria (PK)</font>**
</td>
</tr>
</table>

>* Primary Key
>* Clave primaria
>* Clave principal
>* Llave primaria
>* Llave principal

Una clave primaria en SQL Server es un conjunto de una o más columnas que identifican de manera única cada fila en una tabla.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Características de la Clave Primaria</font>**
</td>
</tr>
</table>

>* **Unicidad**: Los valores en una clave primaria deben ser únicos. No puede haber dos filas con el mismo valor en la columna (o combinación de columnas) que define la clave primaria.
>* **No Nulo**: Las columnas que forman la clave primaria no pueden contener valores nulos. Esto asegura que cada fila pueda ser identificada de manera única.
>* **Índice**: Al definir una clave primaria, SQL Server crea automáticamente un índice único en las columnas de la clave primaria. Esto mejora el rendimiento de las consultas que buscan filas específicas.
>* **Identificación**: La clave primaria es utilizada para establecer relaciones entre tablas. Por ejemplo, puede ser referenciada por claves foráneas en otras tablas, ayudando a mantener la integridad referencial.

## **<font color="#000000"> Identity (Valor_inicial, incremento)</font>**
</td>
</tr>
</table>

El término IDENTITY en SQL Server se utiliza para definir una columna que generará automáticamente valores únicos y secuenciales para cada nuevo registro que se inserte en una tabla.

Esto es especialmente útil para crear claves primarias.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> ¿Como funciona el IDENTITY? </font>**
</td>
</tr>
</table>

>* **Generación Automática**:
Cuando se define una columna como IDENTITY, SQL Server asigna automáticamente un valor a esta columna cada vez que se agrega una nueva fila. No es necesario especificar un valor para esta columna en la instrucción INSERT.
>* **Valores Secuenciales**:
Los valores generados por IDENTITY son secuenciales, lo que significa que el primer valor es el que se define al crear la tabla (por defecto 1) y se incrementa en un valor especificado (por defecto 1) cada vez que se inserta una nueva fila.



```
USE empresa
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in INT IDENTITY
	CONSTRAINT pk_personal_rol_cod_rol_in
	PRIMARY KEY CLUSTERED(cod_rol_in)
	NOT NULL,
	cod_rol_ch CHAR(5),
	nom_rol_vc VARCHAR(20),
)
GO
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0001', 'Ventas');
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0002', 'Almacén');
INSERT INTO Personal.Rol(cod_rol_ch, nom_rol_vc) VALUES ('R0003', 'Personal');
GO
SELECT *
FROM Personal.Rol
GO
```

## **<font color="#000000"> Clave alterna (AK)</font>**
</td>
</tr>
</table>

Una clave alterna es un conjunto de una o más columnas que se utilizan para identificar de manera única cada fila en una tabla, pero que no se designa como la clave primaria.

En otras palabras, una clave alterna es una opción alternativa para identificar filas que también cumple con la propiedad de unicidad.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Características de la Clave Alterna</font>**
</td>
</tr>
</table>

>* **Unicidad**: Al igual que la clave primaria, una clave alterna debe contener valores únicos para cada fila en la tabla. No puede haber duplicados.
>* **No Nulidad**: Las columnas que forman una clave alterna no pueden contener valores nulos, asegurando que cada fila pueda ser identificada de manera única.
>* **Índice Único**: Al definir una clave alterna, SQL Server puede crear un índice único en las columnas que la componen, lo que mejora el rendimiento de las consultas que utilizan estas columnas para buscar datos.
>* **Uso en Relaciones**: Aunque no es la clave primaria, una clave alterna puede ser referenciada por otras tablas, similar a cómo se hace con la clave primaria.



```
USE empresa
GO
IF EXISTS
(
    SELECT *
    FROM sys.schemas s
    INNER JOIN sys.tables t
    ON s.schema_id = t.schema_id
    WHERE
            s.name = 'Personal'
        AND
            t.name = 'Rol'

)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
    cod_rol_in INT IDENTITY
    CONSTRAINT pk_personal_rol_cod_rol_in
    PRIMARY KEY CLUSTERED(cod_rol_in) NOT NULL,
    
	cod_rol_ch CHAR(5) UNIQUE,
    nom_rol_vc VARCHAR(20),
)
GO
SELECT *
FROM Personal.Rol
GO
```

## **<font color="#000000">ÍNDICE (IX)</font>**

</table>

Un índice es una estructura de datos que mejora la velocidad de las operaciones de búsqueda en una tabla a costa de un mayor uso de espacio y un aumento en los tiempos de modificación de datos (inserciones, actualizaciones y eliminaciones).

Los índices funcionan de manera similar a un índice en un libro, que permite localizar información rápidamente sin tener que leer todo el contenido.

<table width='99%'>
<tr>
<td bgcolor='#B00A0F'>

### **<font color="#FFFFFF"> Tipos de Índices</font>**
</td>
</tr>
</table>

>* **Índices Únicos**: Aseguran que los valores en la columna (o columnas) indexadas sean únicos. No permiten duplicados.
>* **Índices No Únicos**: Permiten valores duplicados en la columna indexada.
>* **Índices Clustered**: Determinan el orden físico de los datos en la tabla. Solo puede haber un índice clustered por tabla, y suele ser la clave primaria.
>* **Índices Non-Clustered**: Son índices separados de los datos de la tabla. Pueden existir múltiples índices non-clustered en una tabla.
>* **Índices Compuestos**: Son índices que incluyen más de una columna. Se utilizan para consultas que filtran o buscan en múltiples columnas.



```
USE empresa
GO
IF EXISTS
(
	SELECT name
	FROM sys.indexes
	WHERE name = 'ix_personal_rol_nom_rol_vc'
)
DROP INDEX ix_personal_rol_nom_rol_vc
ON Personal.rol
GO
CREATE INDEX ix_personal_rol_nom_rol_vc
ON Personal.rol(nom_rol_vc)
GO
```
